```python
# CMIP 6 
import os
os.chdir("/home/ybh10/Scripts/")

import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from mpl_toolkits.basemap import Basemap
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pandas as pd
import xarray as xr
from netCDF4 import Dataset
import glob
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from mpl_toolkits.basemap import Basemap
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from scipy.stats import stats
import gc
AOD=[]
import xarray as xr
from my_functions import *
import datetime
```

```python
os.chdir("/home/ybh10/Objective_2/MEDUSA/coupled/Processed/")
Atmos_DMS=xr.open_dataset('Atmos_DMS/daily_dimethyl_sulfide_emission_coupled_jan.nc').m01s50i214*(29/62.13)*1e12
SW_DMS=xr.open_dataset('SW_DMS/daily_mole_concentration_of_dimethyl_sulfide_in_sea_water_coupled_jan.nc').mole_concentration_of_dimethyl_sulfide_in_sea_water*1000000 
cloud_area_frac=xr.open_dataset('daily_cloud_area_fraction_coupled_jan.nc').cloud_area_fraction
chlorophyll=xr.open_dataset('daily_Ocean_Surface_Chlor_coupled_jan.nc').m01s00i096
```

```python
MEDUSA_DMS_std=Atmos_DMS.groupby('time.month').std()
MEDUSA_SWDMS_std=SW_DMS.groupby('time.month').std()
MEDUSA_chlorophyll_std=chlorophyll.groupby('time.month').std()
```

```python
fig = plt.figure(figsize=(10,10), edgecolor='w')
ax = fig.add_subplot(1,1,1)
Atmos_DMS_SO=np.nanmean(Atmos_DMS[:,20:45],axis=(0,2))
MEDUSA_DMS_std_mean=np.nanmean(MEDUSA_DMS_std[0,20:45],axis=(1))

ax.set_ylabel('DMS (ppt)',fontsize=12)
ax.set_xlabel('Latitude',fontsize=12)
ax.grid(linestyle='--',alpha=1)
y_labels = ['64S','62S','60S','58S','56S','54S','52S','50S','48S','46S','44S','42S','40S']

ax.set_title("Observational Atmospheric DMS Feb - Apr",fontsize=20)

ax.errorbar(Atmos_DMS.latitude[20:45],Atmos_DMS_SO,yerr=(MEDUSA_DMS_std_mean),marker='x',color='violet',linestyle='--',linewidth=3,label='HadGEM_CMIP6, x\u0304={:.3f}'.format(np.nanmean(Atmos_DMS_SO)))

plt.legend()
plt.tight_layout()
plt.show()
gc.collect()
```

```python
fig = plt.figure(figsize=(10,10), edgecolor='w')
ax = fig.add_subplot(1,1,1)
SW_DMS_SO=np.nanmean(SW_DMS[:,20:45],axis=(0,2))
MEDUSA_SWDMS_std_mean=np.nanmean(MEDUSA_SWDMS_std[0,20:45],axis=(1))

ax.set_ylabel('DMS (ppt)',fontsize=12)
ax.set_xlabel('Latitude',fontsize=12)
ax.grid(linestyle='--',alpha=1)
y_labels = ['64S','62S','60S','58S','56S','54S','52S','50S','48S','46S','44S','42S','40S']

ax.set_title("Observational Atmospheric DMS Feb - Apr",fontsize=20)

ax.errorbar(Atmos_DMS.latitude[20:45],SW_DMS_SO,yerr=(MEDUSA_SWDMS_std_mean),marker='x',color='violet',linestyle='--',linewidth=3,label='HadGEM_CMIP6, x\u0304={:.3f}'.format(np.nanmean(SW_DMS_SO)))

plt.legend()
plt.tight_layout()
plt.show()
gc.collect()
```
